#APRIORI ALGORITHM

support=0.02, length=2

```python
frequent_itemsets1 = apriori(data, min_support=0.02, use_colnames=True, max_len=2)
frequent_itemsets1
```

confidence=0.7

```python
rules1 = association_rules(frequent_itemsets1, metric="confidence", min_threshold=0.7)
rules1
```
